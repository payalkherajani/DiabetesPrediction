### Exploratory data analysis 

    Dataset - Pima Indians Diabetes
    Source - Kaggle
    
    Performing Exploratory Data Analysis On dataset:
    In exploratory data analysis we try to analyse the data by graphs and statistical concepts
      
    

#### Objective:
This dataset  includes only females who are atleast 21 year old.

We need to predict whether patient has diabetes or not using other features.

###### Importing Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

###### Importing dataset

```python
dia_df = pd.read_csv("diabetes.csv")
```

###### Checking Whether data is balanced or not

```python
dia_df['Outcome'].value_counts()
```

###### Describing each feature

```python
dia_df.describe()
```

Observations:

###### Scatter plot:
       Scatter plots helps us to see relation between any two feature with help of these we can identify most effecting 
       feature in dataset

```python
sns.set_style("whitegrid")
sns.FacetGrid(dia_df , hue = "Outcome" , height = 4).map(plt.scatter , "BloodPressure","BMI").add_legend()
plt.show()
```

OBSERVATION: Data is so much overlapped that it is very difficult and ambiguous to make any prediction from graph

###### Pair Plots

```python
#Pair plot  IT wont work as we have large no of features i.e 9 so we need to reduce dimensionality
plt.close()
sns.set_style("darkgrid")
sns.pairplot(dia_df, hue = "Outcome", height = 6 , vars = ['BloodPressure','BMI','Glucose'])
plt.show()
```

```python
plt.close()
sns.set_style("darkgrid")
sns.pairplot(dia_df, hue = "Outcome", height = 6 , vars = ['Pregnancies','SkinThickness','Insulin','DiabetesPedigreeFunction'])
plt.show()
```

OBSERVATIONS:
    No two attributes are able to clearly separatethe two outcome class instances

###### Checking the Correlation

```python
matrix = np.triu(dia_df.corr())
plt.figure()
sns.heatmap(dia_df.corr(),annot=True,cmap="Oranges",vmin=0,vmax=1,linewidths=3,linecolor='white',fmt='.2g',mask = matrix)
plt.show()
```
